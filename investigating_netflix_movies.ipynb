```python
#importing pands and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import and load data set
netflix_df = pd.read_csv("/content/netflix_data.csv")
```

```python
# Subset the DataFrame not including type "TV Show"
netflix_subset = netflix_df[netflix_df.type != 'TV Show']
netflix_subset.head()
```

```python
# Select specific columns
netflix_movies = netflix_df[['title', 'country', 'genre', 'release_year', 'duration']]
netflix_movies.head()
```

```python
#filter duration less than 60 minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]
short_movies.head(20)
```

```python
# Filter the movies released in the 1990s
df_1990s = netflix_movies[(netflix_movies['release_year'] >= 1990) & (netflix_movies['release_year'] <= 1999)]

# Find the most frequent movie duration
duration = df_1990s['duration'].mode()[0]

# Print the result
print(f"The most frequent movie duration in the 1990s is: {duration} minutes")
```

```python
# Filter the movies released in the 1990s
df_1990s = netflix_movies[(netflix_movies['release_year'] >= 1990) & (netflix_movies['release_year'] <= 1999)]

# Filter short action movies (duration < 90 minutes)
short_action_movies = df_1990s[(df_1990s['duration'] < 90) & (df_1990s['genre'].str.contains('Action', case=False))]

# Count the number of short action movies
short_movie_count = short_action_movies.shape[0]

# Print the result
print(f"The number of short action movies released in the 1990s is: {short_movie_count}")
```

```python
#define empty list
colors = []

#marking non_feature filmand make color list
for index, row in netflix_movies.iterrows():
  if row['genre'] == 'Children':
    colors.append('yellow')
  elif row['genre'] == 'Documentries':
    colors.append('green')
  elif row['genre'] == 'Stand-up':
    colors.append('blue')
  else:
    colors.append('red')

```

```python
#set the fig style
fig = plt.figure(figsize = (12,8))

#create a scatter plot
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors)

#create a title and axis lables
plt.xlabel('Release year')
plt.ylabel('Duration(min)')
plt.title('Movie duration by releasing year')

#show the plot
plt.show()
```

**Conclusion**
Well, as we suspected, non-typical genres such as children's movies are all clustered around the bottom half of the plot. But we can't know for certain until we perform additional analyses.
